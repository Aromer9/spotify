```python
# Se importan las librerías necesarias.
import warnings
warnings.filterwarnings('ignore')
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from tqdm import tqdm
import os
import dotenv
dotenv.load_dotenv()
client_id = os.getenv('client_id')
client_secret = os.getenv('client_secret')
```

```python
# Se crea un objeto de autenticación con las credenciales del cliente.
client_credentials_manager = SpotifyClientCredentials(client_id, client_secret)

sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Se busca información sobre un grupo.
result = sp.search(q='korn', type='artist', limit=1, )
 
# Se imprime el resultado.
result
```

```python
# Se imprime el número de seguidores de un grupo. 
result['artists']['items'][0]['followers']['total']
```

```python
# Se obtiene ID del artista, información antes extraída.
artist = result['artists']['items'][0]['id']
 
# Se obtienen las 10 canciones más escuchadas de Blackpink.
results = sp.artist_top_tracks(artist)
 
# Se imprime el nombre de las canciones y su popularidad
for track in results['tracks']:
    print(f"{track['name']} - Popularidad: {track['popularity']}")
```

```python
def get_artist_by_genre(genre):
    """
    Función que obtiene información de 50 artistas de un género musical
    """
 
    results = sp.search(q=f'genre:{genre}', type='artist', limit=50)
    
    # Extraer información relevante de cada artista
    artist_data = []
    for artist in results['artists']['items']:
        artist_data.append({
            #'id': artist['id'],
            'name': artist['name'],
            'popularity': artist['popularity'],
            'followers': artist['followers']['total'],
            'genres': ', '.join(artist['genres'])
        })
    
    # Convertir la lista en un DataFrame de pandas
    df = pd.DataFrame(artist_data)
    return df
```

```python
genre = 'metal'
df = get_artist_by_genre(genre)
df_sorted = df.sort_values(by='popularity', ascending=False, ignore_index=True)
df_sorted.head(50)
```
